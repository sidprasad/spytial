# Now with Constraints

You can annotate classes with `@cnd` constraints. We still have some figuring out of nominal vs structural selectors.

```python

from cndviz.annotations import cnd

# Define a binary tree node
@cnd('orientation', selector='left', directions=['below', 'left'])
@cnd('orientation', selector='right', directions=['below', 'right'])
class TreeNode:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

    def __repr__(self):
        return f"TreeNode({self.value})"
    
# Create a sample binary tree
root = TreeNode(
    value=10,
    left=TreeNode(
        value=5,
        left=TreeNode(3),
        right=TreeNode(7)
    ),
    right=TreeNode(
        value=15,
        left=TreeNode(12),
        right=TreeNode(18)
    )
)

# Visualize the binary tree
try:
    show(root, method="browser")
except Exception as e:
    print(f"Error visualizing binary tree: {e}")
```
