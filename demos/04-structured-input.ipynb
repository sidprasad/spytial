# Structured input: editing values visually

`spytial.diagram()` *shows* a value. **`spytial.edit()`** is its inverse — it opens that same box‑and‑arrow view as an **editor**: you build or change the value, click **Done**, and it hands the value back to Python as a real object.

- **`edit(obj)`** — opens the editor, **blocks until you click Done**, then **returns a freshly reconstructed object** (no copy‑paste). For a local script or a local Jupyter kernel.
- **`edit_html(obj)`** — a standalone‑HTML editor with an **Export** button; no server needed.

Both accept **any value** — dataclasses, dicts, lists, arbitrary objects — not just dataclasses.

```python
import sys
from pathlib import Path

# Add the parent directory to the Python path
sys.path.append(str(Path().resolve().parent))

from dataclasses import dataclass
from typing import Optional
from spytial import *
```

## Example 1: Simple Binary Tree

Let's start with a classic example: a binary tree node.

We'll define a `TreeNode` dataclass with:
- A `value` field (integer)
- `left` and `right` child references (optional TreeNodes)

Then we'll use `edit_html()` to create an instance visually.

```python
@dataclass
@orientation(selector="{x, y : TreeNode | x.left = y}", directions=["below", "left"])
class TreeNode:
    value: int = 0
    left: Optional['TreeNode'] = None
    right: Optional['TreeNode'] = None
```

```python
t = TreeNode(left=None, right=None, value=17)
edit_html(t)
```

## Example 2: Person with Address

Let's try a more practical example: a Person dataclass with a nested Address.

This demonstrates:
- Multiple dataclass types in one builder
- Nested object references
- Different field types (strings, integers, optional references)

```python
@dataclass
@attribute(field="street")
class Address:
    street: str = ""
    city: str = ""
    zip_code: str = ""

@dataclass
class Person:
    name: str = ""
    age: int = 0
    address: Optional[Address] = None
```

```python
# Start with an empty Person
edit_html(Person())
```

## Example 3: `edit()` — get the value straight back

Examples 1 & 2 use `edit_html()`, which renders standalone HTML you finish with
the built-in **Export** button (copy-paste the constructor code).

`edit()` opens the *same* editor but **blocks until you click Done**, then
reconstructs and **returns the real Python object** directly — no copy-paste.
It serves the editor from a tiny local server (inline in Jupyter, or a browser
tab from a script) and needs only the standard library + a browser.

```python
# Opens the editor and blocks until you click Done, then returns a real TreeNode
result = edit(TreeNode(value=10))
result
```

```python
# `result` is already a real object — no .value step needed
type(result), result
```

### Cancelling

Click **Cancel** (or close the tab / interrupt the kernel) and `edit()` returns
per `on_cancel`: `"seed"` → the original value unchanged (**default**), `"none"`
→ `None`, `"raise"` → raises `EditCancelled`. `edit()` can't hang the kernel — a
closed or unreachable editor unblocks on its own.

```python
# Cancel this one and you get the seed back unchanged
person = edit(Person(name="Alice", age=30), on_cancel="seed")
person
```

### Or raise on cancel

```python
# on_cancel="raise" turns a cancel into an exception you can branch on
try:
    tree = edit(TreeNode(), on_cancel="raise")
    print("got:", tree)
except EditCancelled:
    print("cancelled — nothing returned")
```

## Any value — not just dataclasses

`edit()` accepts whatever `diagram()` accepts. No class definition required — edit a dict, a list, or an arbitrary object and get the same kind back.

```python
# A plain dict — no class needed. Edit it, click Done, get a dict back.
edit({"name": "root", "children": [{"v": 1}, {"v": 2}], "count": 3})
```
