```python
import sys
sys.path.insert(0, '/Users/siddharthaprasad/Desktop/spytial-org/spytial-py')

from dataclasses import dataclass
from typing import Optional
import spytial

@dataclass
@spytial.orientation(selector="left", directions = ["left", "below"])
@spytial.orientation(selector="right", directions = ["right", "below"])
class TreeNode:
    value: int = 0
    left: Optional['TreeNode'] = None
    right: Optional['TreeNode'] = None

builder = spytial.DataClassBuilder(TreeNode(value=42, left=TreeNode(value=21)))
builder
```
